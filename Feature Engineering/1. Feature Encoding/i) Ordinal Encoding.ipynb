## Ordinal Encoding

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"customer.csv")
df
```

```python
df['review'].unique()
```

```python
df['education'].unique()
```

```python
df['purchased'].unique()
```

```python
df['gender'].unique()
```

```python
df['review'] = df['review'].map({'Average':1, 'Poor':0, 'Good':2})
```

```python
df['education'] = df['education'].map({'School':0, 'UG':1, 'PG':2})
```

```python
df['purchased'] = df['purchased'].map({'No':0, 'Yes':1})
```

```python
df['gender'] = df['gender'].map({'Female':0, 'Male':1})
```

```python
df
```

```python
df = pd.read_csv(r"customer.csv")
df
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder()
```

```python
encoded = oe.fit_transform(df[['gender','review','education','purchased']])
encoded
```

```python
encoded_data = pd.DataFrame(encoded,columns=['gender','review','education','purchased'])
encoded_data
```

```python
df.drop(columns=['gender','review','education','purchased'],inplace=True)
```

```python
df = df.join(encoded_data)
df
```

## Label Encoding

```python
df = pd.read_csv(r"customer.csv")
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
lable_df = le.fit_transform(df['review'])
lable_df
```

```python
df['Review_Lable'] = lable_df
df
```

```python
df[['education','gender','purchased']] = df[['education','gender','purchased']].apply(le.fit_transform)
df
```

```python
inverce = le.inverse_transform(df['purchased'])
inverce
```
