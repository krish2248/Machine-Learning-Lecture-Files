# (iii)Robust Scaling

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Sandeep\python lerning\Feature Engineering\2) Feature Scaling\Social_Network_Ads.csv")
df
```

```python
X = df.drop(columns=['User ID','Purchased','Gender'])
X
```

```python
y = df['Purchased']
y
```

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
```

```python
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
X.columns
```

```python
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
X_scaled
```

```python
y
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.scatter(X['Age'], X['EstimatedSalary'],c=y)
plt.title("Before Scaling")

plt.subplot(1,2,2)
plt.scatter(X_scaled['Age'], X_scaled['EstimatedSalary'],c=y)
plt.title("After Scaling")

plt.show()
```
