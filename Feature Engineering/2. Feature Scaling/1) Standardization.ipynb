```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Sandeep\python lerning\Feature Engineering\2) Feature Scaling\Social_Network_Ads.csv")
df.describe()
```

```python
X = df.drop(columns=['User ID','Purchased','Gender'])
X
```

```python
y = df['Purchased']
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
X.columns
```

```python
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
X_scaled
```

```python
round(X.describe(),1)
```

```python
round(X_scaled.describe(),1)
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.scatter(X['Age'], X['EstimatedSalary'],c=y)
plt.title("Before Scaling")

plt.subplot(1,2,2)
plt.scatter(X_scaled['Age'], X_scaled['EstimatedSalary'],c=y)
plt.title("After Scaling")

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(16, 5))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(X['Age'], ax=ax1)             # Kernel Distribution Estimation Plot
sns.kdeplot(X['EstimatedSalary'], ax=ax1)
plt.xlabel('age & salary')

# # after scaling
ax2.set_title('After Standard Scaling')
sns.kdeplot(X_scaled['Age'], ax=ax2)
sns.kdeplot(X_scaled['EstimatedSalary'], ax=ax2)
plt.xlabel('age & salary')

plt.show()
```
