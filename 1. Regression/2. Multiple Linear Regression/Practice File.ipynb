```python
df = pd.read_csv(r"C:\Users\sonik\Desktop\python lerning\ML\1) Regression\2) Multiple Linear Regression\Practical Codes\50_Startups.csv")
```

```python
x = df.drop(columns='Profit')
y = df['Profit']
```

```python
x.head()
```

```python
y.head()
```

```python
df['State'].unique()
```

```python
x = pd.get_dummies(data =x, columns=['State'],drop_first=True)
x.head()
```

```python
from sklearn.preprocessing import RobustScaler

sc = RobustScaler()

x[['R&D Spend', 'Administration', 'Marketing Spend']] = sc.fit_transform(
    x[['R&D Spend', 'Administration', 'Marketing Spend']]
)

x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=102)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model = lr.fit(x_train, y_train)
model
```

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
train_acc = lr.predict(x_train)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
r2_score(y_train, train_acc)
```
