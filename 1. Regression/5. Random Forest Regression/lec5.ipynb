```python
from sklearn.datasets import load_diabetes
data = load_diabetes()
data
```

```python
x = data.data
y = data.target
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split (x,y, test_size=0.4, random_state =42)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators= 100, random_state=42)
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)

y_pred
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_pred, y_test)

print(r2)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
plt.scatter(y_test, y_pred, color='blue', label='Predicted vs Actual')  # blue dots = each prediction
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linewidth=2, label='Perfect Prediction')  # red line = perfect prediction
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Random Forest Regression - Conceptual Visualization")
plt.legend()
plt.show()
```
