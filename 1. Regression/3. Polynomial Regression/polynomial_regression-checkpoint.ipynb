# Polynomial Regression

üîπ Why use it?
Sometimes, data doesn't follow a straight-line trend (like in Linear Regression). Instead, it curves up and down. Polynomial Regression helps us capture these curves.

üîπ How does it work?
Starts with Linear Regression: A simple straight line (y = mx + b).

Adds More Power (Degrees) to X: Instead of just x, we add x¬≤, x¬≥, etc., to create curves.

Example:

Linear: y = a + bx (Straight Line)

Polynomial (2nd degree): y = a + bx + cx¬≤ (Curved Line)

Polynomial (3rd degree): y = a + bx + cx¬≤ + dx¬≥ (More Complex Curve)

Finds the Best Fit: The algorithm adjusts the coefficients (a, b, c, d...) to make the curve fit the data as closely as possible.

üîπ Where is it used?
‚úÖ Stock Market Predictions (Prices go up and down in curves)
‚úÖ Weather Forecasting (Temperature trends can be curved)
‚úÖ Growth Patterns (Like population growth, sales trends)

üîπ Important to Remember
Too simple? Might not capture the trend well.

Too complex? Might overfit (fits data too perfectly but fails on new data).

## Importing the libraries

## Importing the dataset

```python
import pandas as pd

# Load dataset
dataset = pd.read_csv(r'C:\Users\Dell\Downloads\Position_Salaries.csv')

# Extract feature matrix and target variable
X = dataset.iloc[:, 1:-1].values  # Selecting all rows, Level column only
y = dataset.iloc[:, -1].values  # Selecting Salary column
```

## Training the Linear Regression model on the whole dataset

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

## Training the Polynomial Regression model on the whole dataset

This initializes the PolynomialFeatures transformer with a degree of 4, meaning it will generate polynomial terms up to x‚Å¥.

What It Does?
If X originally has one feature (e.g., Level in your dataset), PolynomialFeatures(degree=4) will transform it into multiple features:



```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

## Visualising the Linear Regression results

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Polynomial Regression results

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Polynomial Regression results (for higher resolution and smoother curve)

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
