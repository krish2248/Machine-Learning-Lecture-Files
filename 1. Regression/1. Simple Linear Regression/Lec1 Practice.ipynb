```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
import seaborn as sb
import sklearn
import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv(r"C:\Users\sonik\Desktop\Sample .CSV Data Files\Salary_Data.csv")
dataset
```

```python
x = dataset.drop(columns ='Salary')
y = dataset['Salary']
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
y_predict = regressor.predict(x_test)
y_predict
```

```python
regressor.predict([[10]])
```

```python
x_train_array = np.array(x_train['YearsExperience'])
x_train_array
```

```python
plt.scatter(x_train,y_train,color="g")
plt.plot(x_train_array,regressor.predict(x_train),color='b')
plt.title('Salary vs Experience(training_set)')
plt.xlabel('Years of experience')
plt.ylabel('Salary')
plt.show()
```

```python
x_test_array = np.array(x_test['YearsExperience'])
x_test_array
```

```python
plt.scatter(x_test, y_test, color = 'red')
plt.plot(x_test_array, regressor.predict(x_test), color = 'b')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
mean_absolute_error(y_test, y_predict)
```

```python
r2_score(y_test, y_predict)
```
