# Simple Linear Regression

### Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import pickle
import warnings
```

```python
warnings.filterwarnings("ignore")
```

### importing the dataset

```python
dataset.shape
```

```python
dataset = pd.read_csv("Salary_Data.csv") 
dataset
```

```python
X = dataset.drop(columns='Salary')
y = dataset['Salary']
```

```python
X.head()
```

```python
y.head()
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_test
```

```python
y_test
```

## Training the Simple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

## Predicting the Test set results

```python
y_predict = regressor.predict(X_test)
y_predict
```


## Visualising the Training set results

```python
regressor.predict([[10]])
```

```python
X_train_array = np.array(X_train['YearsExperience'])
X_train_array
```

```python
plt.scatter(X_train,y_train,color="g")
plt.plot(X_train_array,regressor.predict(X_train),color='b')
plt.title('Salary vs Experience(training_set)')
plt.xlabel('Years of experience')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Test set results

```python
X_test_array = np.array(X_test['YearsExperience'])
X_test_array
```

```python
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_test_array, regressor.predict(X_test), color = 'b')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
mean_absolute_error(y_test,y_predict)
```

```python
r2_score(y_test,y_predict)
```
