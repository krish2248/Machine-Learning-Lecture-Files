```python
import pandas as pd #used to mathematical operations on data
import numpy as np #used to cleaning
import matplotlib.pyplot as plt
import pickle #used in multiple
import seaborn as sb #used in multiple
from sklearn.linear_model import LinearRegression #use to import ML models Such as linear regression
import warnings #used in multiple
```

```python
data = {
    'hours' : [1,2,3,4,6,7,8,9,10],
    'marks' : [10,20,30,40,60,70,80,90,100]
}
df = pd.DataFrame(data)
data
```

```python
x = df[['hours']]
y = df['marks']
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
predicted = model.predict([[22]])
print('predicted marks for 10 hours :', predicted[0])
```

```python
plt.scatter(x,y, color = 'red')
plt.plot(x, model.predict(x), color = 'blue')
plt.xlabel('marks')
plt.ylabel('hours')
plt.show()
```
