# 3. Hierarchical Clustering
Hierarchical clustering builds a hierarchy of clusters using a tree-like structure called a **dendrogram**. Unlike K-Means, it does not require the number of clusters (K) to be specified in advance.

## **How It Works:**
1. **Each data point starts as its own cluster.**
2. **Find the two closest clusters and merge them.**
3. **Repeat step 2 until all points form a single cluster.**
4. **The dendrogram helps decide the number of clusters by cutting it at a suitable level.**

## **Types of Hierarchical Clustering:**
- **Agglomerative (Bottom-Up):**
  - Starts with individual points and merges the closest ones.
  - Continues until only one cluster remains.
- **Divisive (Top-Down):**
  - Starts with one large cluster.
  - Recursively splits it into smaller clusters.

## **Real-Life Example:**
Consider how animals are classified in biology:
- **Mammals** – Humans, dogs, and cats.
- **Reptiles** – Snakes, lizards, and turtles.
- **Birds** – Eagles, sparrows, and penguins.

Hierarchical clustering can build a taxonomy tree where closely related species are grouped together.

## **Advantages:**
✅ No need to specify the number of clusters beforehand.  
✅ Produces a tree structure (dendrogram) useful for analysis.  
✅ Works well for small datasets with clear hierarchy.

## **Disadvantages:**
❌ Computationally expensive for large datasets.  
❌ Cannot handle noise and outliers well.  
❌ Difficult to update clusters once the hierarchy is built.

# Using the dendrogram to find the optimal number of clusters

```python
pip install scipy
```

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.datasets import load_iris  # Sample data

# Load sample data
data = load_iris()
X = data.data

# Compute linkage matrix
linked = linkage(X, method='ward')

# Plot dendrogram
plt.figure(figsize=(10, 7))
dendrogram(linked)
plt.title('Dendrogram')
plt.xlabel('Sample index')
plt.ylabel('Euclidean distances')
plt.show()
```

# Training the Hierarchical Clustering model on the dataset

```python
from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='ward')
y_hc = hc.fit_predict(X)
y_hc
```

# Visualising the clusters

```python
plt.scatter(X[y_hc == 0,0], X[y_hc == 0,1],s=100, c='r', label='Cluster 1')
plt.scatter(X[y_hc == 1,0], X[y_hc == 1,1],s=100, c='g', label='Cluster 2')
plt.scatter(X[y_hc == 2,0], X[y_hc == 2,1],s=100, c='b', label='Cluster 3')
plt.scatter(X[y_hc == 3,0], X[y_hc == 3,1],s=100, c='c', label='Cluster 4')
plt.scatter(X[y_hc == 4,0], X[y_hc == 4,1],s=100, c='y', label='Cluster 5')

plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
