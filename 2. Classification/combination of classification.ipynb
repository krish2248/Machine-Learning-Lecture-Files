```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Helly\Desktop\coding\Helly Intern (ML)\7) Machine Learning\2) Classification\1) Logistic Regression\Practical Codes\Social_Network_Ads.csv")
df
```

```python
df.isnull().sum()
```

```python
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# logistic regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,y_train)
```

```python
y_lr = lr.predict(X_test)
```

# K nearest neighbour

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
```

```python
y_knn = knn.predict(X_test)
```

# SVC

```python
from sklearn.svm import SVC
svc = SVC(kernel='linear')
svc.fit(X_train,y_train)
```

```python
y_svc = svc.predict(X_test)
```

```python
from sklearn.svm import SVC
svcr = SVC(kernel='rbf')
svcr.fit(X_train,y_train)
```

```python
y_svcr = svcr.predict(X_test)
```

# naive bayes

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train,y_train)
```

```python
y_nb = nb.predict(X_test)
```

# Decision tree

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(criterion='entropy')
dtc.fit(X_train,y_train)
```

```python
y_dtc = dtc.predict(X_test)
```

# random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(criterion='entropy', n_estimators=30)
rfc.fit(X_train,y_train)
```

```python
y_rfc = rfc.predict(X_test)
```

# evalution

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
cm_lr = confusion_matrix(y_test,y_lr)
cm_knn = confusion_matrix(y_test,y_knn)
cm_svm = confusion_matrix(y_test,y_svc)
cm_ksvm = confusion_matrix(y_test,y_svcr)
cm_nb = confusion_matrix(y_test,y_nb)
cm_dt = confusion_matrix(y_test,y_dtc)
cm_rf = confusion_matrix(y_test,y_rfc)
print('Logiestic Regression:\n',cm_lr)
print('K_nearest neighbour :\n',cm_knn)
print('SVM :\n',cm_svm)
print('KSVM :\n',cm_ksvm)
print('Naive bayes :\n',cm_nb)
print('Decision tree :\n',cm_dt)
print('random Forest :\n',cm_rf)
```

```python
cm_lr = accuracy_score(y_test,y_lr)
cm_knn = accuracy_score(y_test,y_knn)
cm_svm = accuracy_score(y_test,y_svc)
cm_ksvm = accuracy_score(y_test,y_svcr)
cm_nb = accuracy_score(y_test,y_nb)
cm_dt = accuracy_score(y_test,y_dtc)
cm_rf = accuracy_score(y_test,y_rfc)
print('Logiestic Regression:',cm_lr)
print('K_nearest neighbour :',cm_knn)
print('SVM :',cm_svm)
print('KSVM :',cm_ksvm)
print('Naive bayes :',cm_nb)
print('Decision tree :',cm_dt)
print('random Forest :',cm_rf)
```
