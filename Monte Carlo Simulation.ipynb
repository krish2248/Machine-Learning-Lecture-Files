# Crux Of Monte Carlo Simulation 
#### Using random sampling to model uncertainty and estimate outcomes of a complex system or process. 
####                                                              Or In Other Words
#### By using this method one can find the probability of any 2 or more objects or tasks. Giving a uncertain assumed value limit to the objects and running thousands of simulation on it. We can find the probability of the problem very easily and visualise it as required.

```python
import numpy as np
sims = 1000000
a = np.random.uniform(1,5, sims)
b = np.random.uniform(2,6, sims)
duration = a + b
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize= (3, 1.5))
plt.hist(duration, density = True)
plt.axvline(9, color = 'r')
plt.show()
print((duration < 9).sum()/sims)
```
